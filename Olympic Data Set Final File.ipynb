```python
#Imort libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load Data Set
om = pd.read_csv(r"C:\Users\kadah\Downloads\athlete_events.csv")
om2 = pd.read_csv(r"C:\Users\kadah\Downloads\noc_regions.csv")
```

```python
#Top 10 countries participating in olympic
top_10_countries = om_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
plt.figure(figsize=(15,8))
plt.title("Age distribution of the athletes")
plt.xlabel('Age')
plt.ylabel('Number of participants')
plt.hist(om_df.Age, bins = np.arange(8,60,2), color = 'green', edgecolor = 'blue');
```

```python
#Total number of sports played in Winter
winter_sports = om_df[om_df.Season == 'Winter'].Sport.unique()
winter_sports
```

```python
#Total number of sports played in Summer
summer_sports = om_df[om_df.Season == 'Summer'].Sport.unique()
summer_sports
```

```python
plt.figure(figsize=(12,6))
plt.title('top 10 Participating Countries')
sns.barplot(x=top_10_countries.index, y=top_10_countries, palette = 'Set2');
```

```python
#Total number of Male & Female
gender_counts = om_df.Sex.value_counts()
gender_counts
```

```python
plt.figure(figsize = (15,8))
plt.title('Gender Distribution')
plt.pie(gender_counts, labels = gender_counts.index,autopct = '%1.1f%%' , startangle=90, shadow=True);
```

```python
#Total number of Medals won by Country
gm.Region.value_counts().reset_index(name='Medal').head(10)
```

```python
totalGoldMedals = gm.Region.value_counts().reset_index(name='Medal').head(6)
g=sns.catplot(x="index", y="Medal", data=totalGoldMedals,
             height=5,kind='bar',palette="rocket")
g.despine(left=True)
g.set_xlabels("Top 5 cpuntries")
g.set_ylabels("Number of Medals")
plt.title('Gold Medals per Country')
```

```python
#Gold Medals won in year 2016
max_year=om_df.Year.max()
print(max_year)
team_names = om_df[(om_df.Year==max_year) & (om_df.Medal=='Gold')].Team
team_names.value_counts().head(10)
```

```python
sns.barplot(x=team_names.value_counts().head(20), y=team_names.value_counts().head(20).index)
plt.ylabel(None);
plt.xlabel('Countrywise Medals for the year 2016');
```

```python
#Height and Weight of Olympic Medalists
not_null_medals=om_df[(om_df['Height'].notnull()) & (om_df['Weight'].notnull())]
```
